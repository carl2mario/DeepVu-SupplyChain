```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/Apple/Desktop/DataX/Deep Vu project/resupplychainoptimizationdataxprojectteam/metals_daily_train.csv')
data.head()
```

```python
#remove rows where label is missing

mask = data['p1'].isnull()

indexnull = data[mask].index
data = data.drop(index = indexnull, axis =0)
data.head(10)
```

```python
#work with alternate dataframe for generating averages of features for each week
data2 = data.drop(['week_date'], axis=1)


#generate empty matrix to store averages of weeks 
num_rows=len(data2.week.unique()) + 1
num_colns=data2.shape[1]

#the averages matrix will have the average value for each week, i.e. '2008' and each column feature  i.e 'f001_open' 
#the rows are weeks and columns are features 
averages_matrix = np.zeros((num_rows, num_colns))
```

```python

# loop through column names and week names i.e 'f001_open'... and '2008'....
column_names = data2.drop('week',axis=1).columns
week  = data2.week.unique()

#check if there are rows with missing week values
print('any missing value in week column?: ', data2.week.isnull().any())

# to fill matrix, we need the index of each row and column looped through 
# to keep track of column and row number, insert column_number and row_number and increment

# that way we can use the function .loc which takes column names and lets us search easily for NaN instead of .iloc
#also we are not looping through every cell of the dataframe, only on specific weeks and columns a
#and then searching for NaN in a cut-down dataframe

#filling averages matrix 
column_number = 0
for i in column_names:
    row_number = 0
    for j in week:
        average_df = data2.loc[data2.week == j, i].dropna()
        average_guess = average_df.mean()
        
        averages_matrix[row_number, column_number] = average_guess
        row_number = row_number + 1 
    column_number = column_number + 1

print('Estimate Matrix:', averages_matrix)
```

```python
#now we need to fill back the original dataframe i.e. data2 
# the function .loc only fills 1 NaN per loop so the function while had to be added to fill more than 1 value in 
#a certain week with the original average

#also, did not wanted to fill the values where all the rows are NaNs for a certain week 
# but I ended up filling those who had extra values for weekend beyond 5 days

# column_number and row_number for keeping track of number index of weeks and columns looped through to refer to matrix
column_number = 0
for i in column_names:
    row_number=0
    for j in week:
        # keep looping while the number of nulls in smaller dataframe for week i and column j still has NaNs
        # second constraint is to avoid infinite loops if all rows are NaNs 
        
        while( ((data2.loc[ (data2.week == j), i].isnull().sum() ) > 0)\
               & (data2.loc[(data2.week == j), i].isnull().sum() != (data2.loc[(data2.week == j), i].isnull().shape[0])) ):
            
            #fill the dataframe with the values stored in matrix corresponding to week i and column j
            data2.loc[ (data2[i].isnull())&(data2.week == j ), i ] = averages_matrix[row_number,column_number]
        
        row_number = row_number + 1 
    #print to keep track of iteration
    print(i)
    column_number = column_number + 1
```
